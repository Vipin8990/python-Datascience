```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [21,345,56,12,5,47,1]
plt.plot(y)
```

```python
plt.figure(figsize=(15,2))
plt.pie(y)
plt.show()
```

```python
fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(15,5))
plt.show()
```

```python
t = np.arange(-3 * np.pi, 3 * np.pi, 0.2)
x = np.sin(t)
y = np.cos(t)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,2))
ax[0].plot(t,x,'r')
ax[1].bar(t,y)
plt.show()
```
